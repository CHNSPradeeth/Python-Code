```python
import random
print("Rock,Paper and Scissor Game between User and Computer.")
print("Rules for Rock,Paper and Scissor Games:\n1:If Rock vs Paper = Paper.\n2:If Paper vs Scissor = Scissor.\n3:If Rock vs Scissor = Rock.")
class RockPaperScissor:
    def __init__(self):
        self.user_win = 0
        self.computer_win = 0
    def game(self):
        while True:
            print("Your turn:")
            print("Enter the one choice \n1:Rock.\n2:Paper.\n3:Scissor.\n4:Re-Game")
            choice=int(input("Enter your choice:"))
            if choice == 4:
                print("Do you want to play again? (Y/N)")
                ans = input().lower()
                if ans == 'n':
                    break
            while choice>4 or choice<1:
                print("Choose the right choice in between the given list only")
                choice=int(input("Enter the choice present in the above list only:"))
            if choice == 1:
                user_choice = "Rock"
            elif choice == 2:
                user_choice = "Paper"
            else:
                user_choice = "scissor"
            print("Choice of ",choice,"is",user_choice)
            print("Computer turn:")
            computer_choice = random.randint(1,3)
            while computer_choice == choice:
                computer_choice= random.randint(1,3)
            if computer_choice == 1:
                comp_choice = "Rock"
            elif computer_choice == 2:
                comp_choice = "Paper"
            else:
                comp_choice = "Scissor"
            print("Choice of ",computer_choice,"is",comp_choice)
            print(user_choice ,"vs", comp_choice)
            if choice == 1 and computer_choice == 2:
                print("Rock won")
                self.computer_win += 1
            elif choice == 2 and computer_choice == 1:
                print("Rock won")
                self.user_win += 1
            if choice == 1 and computer_choice == 3:
                print("Rock won")
                self.user_win += 1
            elif choice == 3 and computer_choice == 1:
                print("Rock won")
                self.computer_win += 1
            if choice == 2 and computer_choice == 3:
                print("Scissor won")
                self.computer_win += 1
            elif choice == 3 and computer_choice == 2:
                print("Scissor won")
                self.user_win += 1
        print(f"\nPlayer wins: {self.user_win}")
        print(f"Computer wins: {self.computer_win}")
if __name__ == "__main__":
    Games = RockPaperScissor()
    Games.game()
```
