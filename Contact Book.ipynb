```python
# Initialize an empty dictionary to store contacts
contacts_dict = {}

# Infinite loop to continuously show the menu until the user chooses to exit
while True:
    # Display the menu options and take input from the user
    user_choice = int(input("""\t1. Add new contact 2. View contacts 3. Search contact 4. Update contact 5. Delete contact 6. Exit\n"""))
    
    # If user chooses 6, break the loop and exit
    if user_choice == 6:
        print("Exiting")
        break
    
    # Option to add a new contact
    elif user_choice == 1:
        contact_name = input("Enter name:")  # Take the contact's name
        
        # Take input for the number of phone numbers the user wants to save
        phone_limit = int(input("How many phone numbers do you want to save?"))
        
        # Initialize an empty list to store multiple phone numbers
        phone_numbers = []
        
        # Loop to collect phone numbers based on the limit
        while phone_limit > 0:
            phone_number = input("Enter phone number ")  # Ensure contact number is valid (digits and 10 characters)
            if phone_number.isdigit() and len(phone_number) == 10:
                phone_numbers.append(phone_number)  # Add the valid number to the list
            else:
                print("Enter valid number")  # If invalid, prompt again
                continue
            phone_limit -= 1  # Decrement the limit counter
        
        # Collect the email and address for the contact
        email_id = input("Enter email ID")
        contact_address = input("Enter address")
        
        # Store all details in the dictionary with contact_name as the key
        contacts_dict[contact_name] = {
            "Phone Numbers": phone_numbers,
            "Email ID": email_id,
            "Address": contact_address
        }
    
    # If the dictionary is not empty, allow other options
    if contacts_dict != {}:
        
        # Option to view all contacts
        if user_choice == 2:
            # Loop through all contacts and display their details
            for name, details in contacts_dict.items():
                print(f"Name: {name}, Phone numbers: {details['Phone Numbers']}, Email ID: {details['Email ID']}")
        
        # Option to search for a specific contact by name
        elif user_choice == 3:
            search_name = input("Enter name to search ")
            if search_name in contacts_dict:
                # Display the contact's details if found
                contact_details = contacts_dict[search_name]
                print(f"Contact found\nName: {search_name}\nPhone numbers: {contact_details['Phone Numbers']}\nEmail ID: {contact_details['Email ID']}")
            else:
                print("Contact not found")
        
        # Option to update a contact's phone number
        elif user_choice == 4:
            contact_name = input("Enter name")
            if contact_name in contacts_dict:
                contact_details = contacts_dict[contact_name]
                new_phone_number = input("Enter new phone number ")
                # Ensure the new phone number is valid before updating
                if new_phone_number.isdigit() and len(new_phone_number) == 10:
                    contact_details['Phone Numbers'][0] = new_phone_number  # Update the first phone number in the list
                    print("Phone number updated successfully")
            else:
                print("No contact found")
        
        # Option to delete a contact by name
        elif user_choice == 5:
            contact_name = input("Enter name to delete contact")
            if contact_name in contacts_dict:
                del contacts_dict[contact_name]  # Delete the contact from the dictionary
                print("Deleted successfully")
            else:
                print("Contact not found")

    # If the dictionary is empty, prompt the user that no contacts exist
    else:
        print("Dictionary is empty")
        continue
```
